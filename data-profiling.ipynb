```python
import os

import pandas as pd
from PIL import Image
from ydata_profiling import ProfileReport
```

```python
from PIL import Image
from PIL.ExifTags import TAGS

def analyze_exif(image_path):
    try:
        image = Image.open(image_path)
        exif_data = image.getexif()

        if exif_data:
            extracted_data = {}

            for k, v in exif_data.items():
                tag = TAGS.get(k),
                extracted_data[tag] = v
            # extracted_data = {
            #     "Camera Model": exif_data.get(37890),
            #     "Camera Model": exif_data.get(37890),  # MakerNote - Camera Model
            #     "Exposure Time": exif_data.get(33434),  # Exposure Time (reciprocal)
            #     "F-Number": exif_data.get(37378),      # FNumber
            #     "Focal Length": exif_data.get(37386),  # FocalLength
            #     "Date Taken": exif_data.get(36867),     # DateTimeOriginal
            # }
            return extracted_data
        else:
            return {"EXIF Data": "Not Available"}
    except (IOError, OSError):
        return {"Error": f"Failed to open image: {image_path}"}
```

```python
image_data = []

for image in os.listdir("images"):
    if image.lower().endswith((".jpg", ".jpeg", ".png", "tif", "tiff")):
        image_path = os.path.join("images", image)
        exif_info = analyze_exif(image_path)
        image_data.append(exif_info)

df = pd.DataFrame(image_data)
df.head()
```

```python
profile = ProfileReport(df, title="EXIF Metadata Report")
profile.to_file("profile.html")
```
